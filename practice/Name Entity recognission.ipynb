```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.chunk import ne_chunk # finding name and entity of words
```

```python
words = word_tokenize(sentence)
tag_elements = nltk.pos_tag(words)
ne_chunk(tag_elements)
```
